#### 3.- Para el encabezado, extrae empleando expresiones regulares el identificador. (1pto)

```python
import re
from Bio import SeqIO

# Leer archivo FASTA
record = SeqIO.read("genFASTA.fasta", "fasta")

# Guardar encabezado en variable
encabezado = record.description.replace("\n", "")

# Usar expresión regular para capturar el identificador al inicio
# ^\S+  => "desde el inicio de la cadena, toma la primera palabra sin espacios"
match = re.match(r"^\S+", encabezado)

if match:
    identificador = match.group(0)
else:
    identificador = None

# Mostrar resultados
print("Encabezado completo:", encabezado)
print("Identificador extraído:", identificador)
```

#### 5.- Devuelve la secuencia de ARN que surgiría a partir de la siguiente secuencia. Recuerda que la secuencia de ARN resultante tiene que ser complementaria a la de ADN. (1 pto)

```python
def dna_to_rna_complement(dna_seq):
    # Diccionario de complementariedad ADN -> ARN
    complement = {
        "A": "U",  # A en ADN -> U en ARN
        "T": "A",  # T en ADN -> A en ARN
        "G": "C",  # G en ADN -> C en ARN
        "C": "G"   # C en ADN -> G en ARN
    }
    # Generar la secuencia complementaria de ARN
    rna_seq = "".join(complement[base] for base in dna_seq)
    return rna_seq

# Aplicar la función a tu secuencia del gen CNTNAP2
rna = dna_to_rna_complement(secuencia)

# Mostrar resultados
print("Encabezado:", encabezado)
print("\nPrimeros 200 nucleótidos de la secuencia de ARN:")
print(rna[:200])
print("\nLongitud de la secuencia de ARN:", len(rna), "nucleótidos")
```

#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}
```

```python
def digestion(seq, enzimas):
    enz_cortan = {}
    enz_no = []
    cortes = []

    # Buscar sitios de corte
    for enz, sitio in enzimas.items():
        sitio_sin = sitio.replace("*", "")
        pos_rel = sitio.index("*")
        pos_encontradas = []

        start = 0
        while True:
            idx = seq.find(sitio_sin, start)
            if idx == -1:
                break
            pos_corte = idx + pos_rel
            pos_encontradas.append(pos_corte)
            cortes.append(pos_corte)
            start = idx + 1
        
        if pos_encontradas:
            enz_cortan[enz] = pos_encontradas
        else:
            enz_no.append(enz)
    
    # Calcular fragmentos
    cortes = sorted(set(cortes))
    fragmentos = []
    prev = 0
    for c in cortes:
        fragmentos.append(c - prev)
        prev = c
    fragmentos.append(len(seq) - prev)

    # Mostrar resultados
    print("Enzimas que cortan:")
    for enz, pos in enz_cortan.items():
        print(f" - {enz}: {pos}")
    print("\nEnzimas que no cortan:", enz_no)
    print("\nFragmentos resultantes:", fragmentos)
    print("Número de fragmentos:", len(fragmentos))
```

```python
from Bio import SeqIO

record = SeqIO.read("genFASTA.fasta", "fasta")
secuencia = str(record.seq)
print("Longitud de la secuencia:", len(secuencia))
```

```python
digestion(secuencia, enzR)
```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
from Bio.Seq import Seq

def ajustar_seq(seq):
    """Recorta la secuencia para que su longitud sea múltiplo de 3"""
    resto = len(seq) % 3
    if resto != 0:
        seq = seq[:-resto]  # elimina los últimos 1 o 2 nucleótidos
    return seq

def orfs_simple(seq_dna):
    # Ajustar longitud
    seq_dna = ajustar_seq(seq_dna)
    seq_dna = Seq(seq_dna)
    orfs = []

    # Hebra positiva
    for frame in range(3):
        prot = seq_dna[frame:].translate(to_stop=False)
        for p in prot.split("*"):
            if len(p) > 0:
                orfs.append((p, "+"))

    # Hebra negativa
    seq_neg = seq_dna.reverse_complement()
    for frame in range(3):
        prot = seq_neg[frame:].translate(to_stop=False)
        for p in prot.split("*"):
            if len(p) > 0:
                orfs.append((p, "-"))

    # Ordenar por tamaño
    orfs = sorted(orfs, key=lambda x: len(x[0]), reverse=True)

    # 6 ORFs más largos
    top6 = orfs[:6]

    # Mejor ORF (más largo)
    best_orf, strand = orfs[0]

    # % aromáticos
    aromaticos = sum(best_orf.count(aa) for aa in "FWY")
    porcentaje_arom = (aromaticos / len(best_orf)) * 100 if len(best_orf) > 0 else 0

    # Longitud relativa
    rel = (len(best_orf) * 3 / len(seq_dna)) * 100

    # Imprimir resultados
    print("6 ORFs más largos:")
    for i, (orf, strand) in enumerate(top6, 1):
        print(f"{i}. Longitud {len(orf)} aa, Hebra {strand}")

    print("\nMejor ORF (más largo):")
    print(f"Longitud {len(best_orf)} aa, Hebra {strand}")

    print("\nAnálisis de la proteína predicha:")
    print(f"% aromáticos: {porcentaje_arom:.2f}%")
    print(f"Longitud relativa al ADN: {rel:.2f}%")

    # 🔹 Devolver resultados
    return {
        "top6": [(len(o), s) for o, s in top6],
        "best_orf_seq": best_orf,
        "best_orf_len": len(best_orf),
        "strand": strand,
        "aromatic_percent": porcentaje_arom,
        "relative_length": rel
    }
```
